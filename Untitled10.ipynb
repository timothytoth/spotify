```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv(r"C:\Users\Tim\OneDrive\Desktop\streaming_history")
```

```python
df['artistName'] = df['artistName'].str.lower()
df['trackName'] = df['trackName'].str.lower()
```

```python
df.head(50)
```

```python
#Calculate the total play count for each unique combination of 'artistName' and 'trackName'
total_count = df.groupby(['artistName', 'trackName'])['msPlayed'].count().reset_index()
total_count.columns = ['artistName', 'trackName', 'totalCount']

# Merge the total play count ('totalCount') back into the DataFrame
df = df.merge(total_count, on=['artistName', 'trackName'], how='left')

#Sort the DataFrame by 'endTime' in descending order
df = df.sort_values(by='endTime', ascending=False)

#Keep only the first row in each group (latest date played)
df = df.drop_duplicates(subset=['artistName', 'trackName'], keep='first')

# Now df contains the songs with the latest date played, aggregated msPlayed, and totalCount
```

```python
df.head(50)
```

```python
df = df.drop(columns=[ 'Unnamed: 0'])
```

```python
df['trackName'] = df['trackName'].str.replace(r'["().*\/-:]', '', regex=True)
```

```python
df.shape
```

```python
#Group by 'artistName' and calculate the total 'totalCount' for each artist
artist_totals = df.groupby('artistName')['totalCount'].sum().reset_index()

#Sort the artists by 'totalCount' in descending order
artist_totals = artist_totals.sort_values(by='totalCount', ascending=False)

# Get the top 10 artists
top_10_artists = artist_totals.head(10)

#Create a bar plot to visualize the top 10 artists by 'totalCount'
plt.figure(figsize=(12, 6))
plt.bar(top_10_artists['artistName'], top_10_artists['totalCount'])
plt.xlabel('Artist Name')
plt.ylabel('Total Count')
plt.title('Top 10 Artists by Total Count')
plt.xticks(rotation=45)
plt.show()
```
